```python
import matplotlib.pyplot as plt
```

```python
def get_plot_bar(df, input_name, title):
  df_indexed = df.set_index('week_ini_date')
  # to set the plot size
  plt.figure(figsize=(16, 8), dpi=150)

  # using plot method to plot open prices.
  # in plot method we set the label and color of the curve.
  df_indexed[f'{input_name}_se'].plot(label=f'{input_name}_se', color='orange', kind='bar',stacked=True)
  df_indexed[f'{input_name}_s'].plot(label=f'{input_name}_s', color='blue',kind='bar',stacked=True)
  df_indexed[f'{input_name}_ne'].plot(label=f'{input_name}_ne', color='green',kind='bar',stacked=True)
  df_indexed[f'{input_name}_n'].plot(label=f'{input_name}_n', color='red',kind='bar',stacked=True)

  # adding title to the plot
  plt.title(f'{title}')

  # adding Label to the x-axis
  plt.xlabel('Weekly Observations')

  # adding legend to the curve
  plt.legend()
```

```python
get_plot_bar(df_exchanges, 'exchange','Energy Exchange Net (MWmed)')
```

```python
df_benchmark_se = pd.read_excel('benchmark_consolidated.xlsx', sheet_name='southeast')
df_benchmark_ne = pd.read_excel('benchmark_consolidated.xlsx', sheet_name='northeast')
df_benchmark_s = pd.read_excel('benchmark_consolidated.xlsx', sheet_name='south')
df_benchmark_n = pd.read_excel('benchmark_consolidated.xlsx', sheet_name='north')
df_benchmark_se
```

```python
def trend_direction_accuracy(df_benchmark):
  total = 0
  actual_up = 0
  actual_down = 0
  lstm_up = 0
  lstm_down = 0
  decomp_up = 0
  decomp_down = 0
  stable = 0
  predictions_right_direction = 0
  decomp_right_direction = 0
  for index in df_benchmark.index:
    if index == 0:
      pass
    else:
      if df_benchmark.at[index, 'actuals'] > df_benchmark.at[index-1, 'actuals']:
        actual_up = actual_up + 1
        if df_benchmark.at[index, 'predictions'] > df_benchmark.at[index-1, 'predictions']:
          predictions_right_direction = predictions_right_direction + 1
          lstm_up = lstm_up + 1
        if df_benchmark.at[index, 'decomp_predictions_as_of'] > df_benchmark.at[index-1, 'decomp_predictions_as_of']:
          decomp_right_direction = decomp_right_direction + 1
          decomp_up = decomp_up + 1
      elif df_benchmark.at[index, 'actuals'] < df_benchmark.at[index-1, 'actuals']:
        actual_down = actual_down + 1
        if df_benchmark.at[index, 'predictions'] < df_benchmark.at[index-1, 'predictions']:
          predictions_right_direction = predictions_right_direction + 1
          lstm_down = lstm_down + 1
        if df_benchmark.at[index, 'decomp_predictions_as_of'] < df_benchmark.at[index-1, 'decomp_predictions_as_of']:
          decomp_right_direction = decomp_right_direction + 1
          decomp_down = decomp_down + 1
      else:
        stable = stable + 1
        # if df_benchmark.at[index, 'predictions'] == df_benchmark.at[index-1, 'predictions']:
        #   predictions_right_direction = predictions_right_direction + 1
        # if df_benchmark.at[index, 'decomp_predictions_as_of'] == df_benchmark.at[index-1, 'decomp_predictions_as_of']:
        #   decomp_right_direction = decomp_right_direction + 1
      total = total + 1
  total = total - stable
  print(f'Total trends considered: {total}')
  print(f'Actual UP trends: {actual_up}')
  print(f'Actual DOWN trends: {actual_down}')
  print(f'LSTM UP trends: {lstm_up}')
  print(f'LSTM DOWN trends: {lstm_down}')
  print(f'DECOMP UP trends: {decomp_up}')
  print(f'DECOMP DOWN trends: {decomp_down}')
  print(f'LSTM Trend Direction Accuracy: {(predictions_right_direction/(total))*100}')
  print(f'DECOMP Trend Direction Accuracy: {(decomp_right_direction/(total))*100}')
```

```python
trend_direction_accuracy(df_benchmark_se)
trend_direction_accuracy(df_benchmark_ne)
trend_direction_accuracy(df_benchmark_s)
trend_direction_accuracy(df_benchmark_n)
```

```python
def get_rmse_vol(df_benchmark):
  for index in df_benchmark.index:
    if index == 0:
      df_benchmark.at[index, 'actual_vol'] = 0
      df_benchmark.at[index, 'lstm_vol'] = 0
      df_benchmark.at[index, 'decomp_vol'] = 0
    else:
      df_benchmark.at[index, 'actual_vol'] = ((df_benchmark.at[index, 'actuals'] - df_benchmark.at[index-1, 'actuals'])/df_benchmark.at[index-1, 'actuals'])*100
      df_benchmark.at[index, 'lstm_vol'] = ((df_benchmark.at[index, 'predictions'] - df_benchmark.at[index-1, 'predictions'])/df_benchmark.at[index-1, 'predictions'])*100
      df_benchmark.at[index, 'decomp_vol'] = ((df_benchmark.at[index, 'decomp_predictions_as_of'] - df_benchmark.at[index-1, 'decomp_predictions_as_of'])/df_benchmark.at[index-1, 'decomp_predictions_as_of'])*100
  rmse_lstm_vol = sqrt(mse(df_benchmark['actual_vol'], df_benchmark['lstm_vol']))
  print('RMSE LSTM Vol: %.3f' % rmse_lstm_vol)
  rmse_decomp_vol = sqrt(mse(df_benchmark['actual_vol'], df_benchmark['decomp_vol']))
  print('RMSE DECOMP Vol: %.3f' % rmse_decomp_vol)
```

```python
get_rmse_vol(df_benchmark_se)
get_rmse_vol(df_benchmark_ne)
get_rmse_vol(df_benchmark_s)
get_rmse_vol(df_benchmark_n)
```

```python
def get_rmse_absolute(df_benchmark):
  rmse_lstm = sqrt(mse(df_benchmark['actuals'], df_benchmark['predictions']))
  print('RMSE LSTM Absolute Values: %.3f' % rmse_lstm)
  rmse_decomp = sqrt(mse(df_benchmark['actuals'], df_benchmark['decomp_predictions_as_of']))
  print('RMSE DECOMP Absolute Values: %.3f' % rmse_decomp)
```
